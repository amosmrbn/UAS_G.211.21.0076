```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset curah hujan dari file CSV
df = pd.read_csv('curah_hujan.csv')

# Visualisasi menggunakan seaborn
sns.set(style="whitegrid")

# Membuat plot bar
plt.figure(figsize=(12, 7))
ax = sns.barplot(x='Bulan', y='Curah_Hujan', hue='Tahun', data=df, palette=['blue', 'red', 'green'])
plt.xlabel('Bulan')
plt.ylabel('Curah Hujan (mm)')
plt.ylim(0, 700)
legend = plt.legend(labels=['1981-1990', '1991-2000', '2001-2010'])
legend.get_texts()[0].set_color('blue')
legend.get_texts()[1].set_color('red')
legend.get_texts()[2].set_color('green')
plt.show()
```
